```r
# cast to int
age <- "20"
flag <- FALSE
as.numeric(age)
as.numeric(flag)
print(age) # 20
print(flag) # 0
# cast to str
age <- 20
flag <- 0
as.character(age)
as.character(flag)
print(age) # "20"
print(flag) # "FALSE"

# cast to boolean
flagOne <- 1
flagTwo <- 0
as.logical(flagOne)
as.logical(flagTwo)
print(flagOne) # FALSE
print(flagTwo) # FALSE
typeof(flagOne)
class(flagOne)
```

the typeof() helps understand the specific data types of their objects, like we can see that flagOne was indeed 'numeric' but it was actually 'double'

```r
myPower <- function(b, e) {
  # result to 1
  result <- 1
  if(e == 0){
    return(result)
  }

  # negative exponents
  if (e < 0) {
    b <- 1 / b
    e <- -e
  }

  # power using a loop
  for (i in 1:e) {
    result <- result * b
  }

  # Return the result
  return(result)
}

print(myPower(2, 3))  # Output: 8
print(myPower(3, 4))  # Output: 81
print(myPower(5, 0))  # Output: 1
print(myPower(-1, -1)) # Output: -1
```

```r
subtotal <- function(values) {
  cumulative_sum <- values[1]

  # get cumsum
  for (i in 2:length(values)) {
    cumulative_sum <- cumulative_sum + values[i]
    values[i] <- cumulative_sum
  }

  # return modified array
  return(values)
}

arr <- c(5.8, 2.6, 9.1, 3.4, 7.0)
values <- subtotal(arr)

# Print the modified array and the average
print(arr)
print(values)
print(mean(values))
```

```r
Counter <- function(char_array) {
  letters <- 0
  uppercases <- 0
  lowercases <- 0
  spaces <- 0


  for (i in seq_along(char_array)) {
    if (grepl("[A-Za-z]", char_array[i])) { # grepl is builtin pattern matching (regex)
      letters <- letters + 1
    }
    if (grepl("[A-Z]", char_array[i])) {
      uppercases <- uppercases + 1
    } else if (grepl("[a-z]", char_array[i])) {
      lowercases <- lowercases + 1
    } else if (char_array[i] == " ") {
      spaces <- spaces + 1
    }
  }

  return(c(letters, uppercases, lowercases, spaces))
}


text <- "Hello, I am a student of FAST University."
# string into a character vector
char_array <- unlist(strsplit(text, split = ""))
result <- Counter(char_array)

cat(text, "\n")
cat('No. of letters are:', result[1], "\n")
cat('No. of spaces are:', result[4], "\n")
cat('No. of uppercase letters are:', result[2], "\n")
cat('No. of lowercase letters:', result[3], "\n")
```

```r
findMedianSortedArrays <- function(nums1, nums2) {
    # extend nums1 with nums2
    nums1 <- c(nums1, nums2)
    # Sort array
    nums1 <- sort(nums1)
    # if length is even
    if (length(nums1) %% 2 == 0) {
        # If even, return average middle two elements
        return((nums1[length(nums1) / 2] + nums1[length(nums1) / 2 + 1]) / 2)
    } else {
        # If odd, return middle element
        return(nums1[(length(nums1) + 1) / 2])
    }
}


print(findMedianSortedArrays(c(1, 3), c(2)))
print(findMedianSortedArrays(c(1, 2), c(3, 4)))
```
