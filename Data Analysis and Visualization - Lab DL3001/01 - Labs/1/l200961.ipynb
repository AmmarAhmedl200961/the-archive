```python
# 1
# number of rows for the pattern
rows = 4

# iterate through each row
for i in range(rows):
    # '*' for the first and last row
    if i == 0 or i == rows - 1:
        print("*" * rows)
    else:
        # '**' only for the middle rows
        print("**")
```

```python
# 3
def factorial(n):

  # single line to find factorial
  return 1 if (n==1 or n==0) else n * factorial(n - 1)
# Driver Code
num = 6
print("Factorial of",num,"is",factorial(num))
```

```python
# 4
# Reversing a list using slicing technique
def Reverse(lst):
  return lst[::-1]
lst = [10, 11, 12, 13, 14, 15]
print(Reverse(lst))
```

```python
# 5
from functools import reduce

def pro_su(arr):
    pro = reduce((lambda x, y: x * y), arr)
    su = reduce((lambda x, y: x + y), arr)
    return pro, su
lst = [1, 2, 3, 4, 5]
print(pro_su(lst))
```

```python
# 6
def copy_list(lst1, lst2):
  for item in lst1:
    lst2.append(item)
lst1 = [1, 2, 3, 4, 5]
lst2 = []
copy_list(lst1, lst2)
print(lst2)
```

```python
# 7
def count_frequency(lst):
  freq = {}
  for item in lst:
    if isinstance(item, int): # check only int datatype as list can house other datatypes
      if item in freq:
        freq[item] += 1
      else:
        freq[item] = 1
  return freq

items = [1, 1, 2, 3, 3, 4, 5, 5]
print(count_frequency(items))
```

```python
# 8
def find_max_min(tup):
  return max(tup), min(tup)

# Example usage
my_tuple = (10, 5, 20, 1, 30)
max_value, min_value = find_max_min(my_tuple)

print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

```python
# 9
def is_palindrome(lst):
  return lst == lst[::-1]

# Example usage
my_list1 = [1, 2, 3, 2, 1]
my_list2 = [1, 2, 3, 4, 5]

print(is_palindrome(my_list1))  # Output: True
print(is_palindrome(my_list2))  # Output: False
```
