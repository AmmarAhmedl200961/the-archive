# **Question 1:** You are analyzing monthly sales data for a store and have the following values:

- Months: ['January', 'February', 'March', 'April']
- Sales: [5000, 7000, 8000, 6000]


Create a line chart using Plotly to visualize this data.

````python
import plotly.express as px
# Sample data
months = ['January', 'February', 'March', 'April']
sales = [5000, 7000, 8000, 6000]

# Create a line chart
fig = px.line(x=months, y=sales, title='Monthly Sales')

# Customize the chart (optional)
fig.update_xaxes(title_text='Month')
fig.update_yaxes(title_text='Sales')

# Show the chart
fig.show()
````

# Question 2: Interactive Scatter Plot with Labels

A researcher collects data for a study on plant growth. The heights (in cm) and widths (in cm) of 6 plants are:

- Heights: [10, 15, 20, 25, 30, 35]
- Widths: [3, 4, 5, 6, 7, 8]

Plot an interactive scatter plot and label each point with "Plant 1", "Plant 2", and so on.

````python
import plotly.graph_objects as go

heights = [10, 15, 20, 25, 30, 35]
widths = [3, 4, 5, 6, 7, 8]

fig = go.Figure()

for i in range(len(heights)):
  fig.add_trace(go.Scatter(
      x=[heights[i]],
      y=[widths[i]],
      mode='markers+text',
      text=['Plant ' + str(i+1)],
      textposition='top center',
      marker=dict(size=10)
  ))

fig.update_layout(title='Plant Growth Data',
                  xaxis_title='Height (cm)',
                  yaxis_title='Width (cm)')

fig.show()
````

 # Question 3:  Bar Chart for Product Sales
A company sells four products: A, B, C, and D. The sales in Q1 are as follows:

- Products: ['A', 'B', 'C', 'D']
- Sales: [150, 250, 300, 200]

Create a bar chart and customize it by giving the bars different colors.

````python
products = ['A', 'B', 'C', 'D']
sales = [150, 250, 300, 200]
colors = ['red', 'green', 'blue', 'orange']

fig = go.Figure(data=[go.Bar(
    x=products,
    y=sales,
    marker_color=colors
)])

fig.update_layout(title='Product Sales in Q1',
                  xaxis_title='Products',
                  yaxis_title='Sales')

fig.show()
````

 # Question 5: Add Dropdown for Data Selection
Create a line chart for temperature data for two cities (City A and City B). Allow users to select which city's data they want to see using a dropdown.

- City A: [10, 12, 15, 18]
- City B: [8, 9, 11, 14]
- Months: ['January', 'February', 'March', 'April']



````python
# data
months = ['January', 'February', 'March', 'April']
city_a_temps = [10, 12, 15, 18]
city_b_temps = [8, 9, 11, 14]

# initialize figure
fig = go.Figure()

# traces for each city
fig.add_trace(go.Scatter(x=months, y=city_a_temps, mode='lines', name='City A', visible=True))
fig.add_trace(go.Scatter(x=months, y=city_b_temps, mode='lines', name='City B', visible=False))


# dropdown buttons
buttons = [
    dict(label='City A',
         method='update',
         args=[{'visible': [True, False]},
               {'title': 'Temperature Data for City A'}]),
    dict(label='City B',
         method='update',
         args=[{'visible': [False, True]},
               {'title': 'Temperature Data for City B'}]),
    dict(label='Both',
         method='update',
         args=[{'visible': [True, True]},
               {'title': 'Temperature Data for Both Cities'}]),
]

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=buttons,
            x=0.5,  # position of the dropdown
            xanchor='center',
            y=1.15,
            yanchor='top'
        )
    ],
    title='Temperature Data'
)

# Show the chart
fig.show()
````

 # Question 6:

Plot the following points in a 3D space:

- x-coordinates: [1, 2, 3, 4, 5]
- y-coordinates: [2, 4, 6, 8, 10]
- z-coordinates: [1, 3, 5, 7, 9]

````python
# data for the 3D plot
x_coords = [1, 2, 3, 4, 5]
y_coords = [2, 4, 6, 8, 10]
z_coords = [1, 3, 5, 7, 9]

fig = px.scatter_3d(x=x_coords, y=y_coords, z=z_coords, title='3D Scatter Plot')

fig.update_layout(scene=dict(xaxis_title='X-coordinate',
                             yaxis_title='Y-coordinate',
                             zaxis_title='Z-coordinate'))

fig.show()
````

 # Question 7:  Annotate Data Points

- Math Scores: [78, 85, 90, 95, 88]
- Science Scores: [80, 70, 92, 85, 86]

Create a scatter plot and annotate each point with the student names.

# Question 9: Dual Axis Chart
Create a chart that visualizes the monthly rainfall and temperature for a city:

- Months: ['Jan', 'Feb', 'Mar', 'Apr']
- Rainfall (mm): [20, 30, 50, 40]
- Temperature (C): [5, 7, 10, 12]

Use dual axes, one for rainfall and one for temperature.

````python
months = ['Jan', 'Feb', 'Mar', 'Apr']
rainfall = [20, 30, 50, 40]
temperature = [5, 7, 10, 12]

fig = go.Figure()

fig.add_trace(go.Bar(x=months, y=rainfall, name='Rainfall (mm)'))
fig.add_trace(go.Scatter(x=months, y=temperature, name='Temperature (C)', yaxis='y2'))

fig.update_layout(
    title='Monthly Rainfall and Temperature',
    yaxis=dict(title='Rainfall (mm)'),
    yaxis2=dict(title='Temperature (C)', overlaying='y', side='right')
)

fig.show()
````

# Question 10:
A teacher records scores for 3 subjects (Math, Science, English) for 4 students:

- Students: ['Alice', 'Bob', 'Charlie', 'David']
- Subjects: ['Math', 'Science', 'English']

```
Scores: [[78, 80, 85],  
         [90, 70, 88],  
         [85, 92, 79],  
         [88, 85, 84]]
```

Create a heatmap to visualize this data.




````python
students = ['Alice', 'Bob', 'Charlie', 'David']
subjects = ['Math', 'Science', 'English']
scores = [[78, 80, 85],
         [90, 70, 88],
         [85, 92, 79],
         [88, 85, 84]]

fig = px.imshow(scores,
                labels=dict(x="Subjects", y="Students", color="Scores"),
                x=subjects,
                y=students,
                title="Student Scores Heatmap")
fig.show()
````
