```python
def to_lower(sample):
    # complete your function which returns
    # sample lowered case
    for i in range(len(sample)):
        sample1=sample.lower()
        return sample1

```

```python
def remove_stop_words(sample, stop_words):
    # complete your function which
    # can return sample without stop words
    words = sample.split()  # sample into words
    filtered_words = [word for word in words if word not in stop_words]
    result = ' '.join(filtered_words)
    return result



```

```python
def normalize_data(sample, stop_words):
    # first make it lower case
    # then remove stop words
    # also remove symbols, like [], () etc
    symbols=["(",")",",","-","."]

    x=remove_stop_words(sample, stop_words)
    x = x.split()
    for word in symbols:
        if word in x:
            x = [w for w in x if w != word]
    return ' '.join(x)
```

```python
def unique_words(sample):
   # returns list of all unique words
   # word ends at a space
   sample=sample.split()
   unique=list(set(sample))
   return unique
```

```python
def bag_of_words(sample_list):
   # store all unique words in all samples of text
   # sample1 = sample_list[0]
   # return unique_words(sample1) + unique_words(sample2)
   unq=sample_list[0]+sample_list[1]+sample_list[2]

   u=list(set(unq))
   return u
```

```python
def tf_idf(sample_list, bagOfWords):
  # dfi = number of times a 'term' occured in all samples
  # W_ij = tf_i * log(D/dfi)
  # D: number of document/samples
  # return a matrix W_ij
  # where matrix has [Q,D1,D2,D3....]
  # refer to the given iamges above
  D=[]

  for j in bagOfWords:
    count=0
    for i in sample_list:
      if(j==i):
        count = count+1
    D.append(count)
  return D



```

```python
def cosin_similarity(v1,v2):
    # cosine similarity of vector1 to vector2

    dot_product = sum(a * b for a, b in zip(v1, v2))
    magnitude_v1 = sum(a * a for a in v1) ** 0.5
    magnitude_v2 = sum(b * b for b in v2) ** 0.5
    if magnitude_v1 == 0 or magnitude_v2 == 0:
        return 0
    return dot_product / (magnitude_v1 * magnitude_v2)
```

```python
def rank(W_ij):
    # where W_ij = tf_idf(....)
    # Q is at first position
    # take dot product of Q and W_ij[i+1]
    # now reduce all entries using sqrt(i^2 + j^2 ...)
    # put result in dictionary, using key=index of sample
    # sort the result based on reduced value
    # return rank_dic
    Q = W_ij[0]
    similarities = {}

    for i in range(1, len(W_ij)):
        similarity = cosin_similarity(Q, W_ij[i])
        similarities[i] = similarity

    sorted_similarities = dict(sorted(similarities.items(), key=lambda item: item[1], reverse=True))

    return sorted_similarities
```
