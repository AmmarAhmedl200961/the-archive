**Part I: Find the matching substrings (Marks=3)** <br>


Write a function get_substring_positions() that accepts two strings as arguments. Return the number of the positions where they contain the same substring of length 2. For example, **"kerase"** and **"ketasee"** should yield 3, since the "ke", "as", and "se" substrings appear in the same place in both strings.

```python
def get_substring_positions(s1, s2):
    count = 0
    for i in range(len(s1)-1):
        # check substring of length 2
        if s1[i:i+2] == s2[i:i+2]:
            count += 1
    return count

print(get_substring_positions('kerase','ketasee'))
```

**A: Find the most expensive car company name (Marks= 2)**

```python
import pandas as pd

df = pd.read_csv('Automobile_data.csv')
# find most expensive car company name
print(df[df['price'] == df['price'].max()]['company'])
```

**B: Count the total cars per company (Marks=2)**

```python
# count total cars per company
print(df['company'].value_counts())
```

**Part C: List the names of top 10 highest price cars of *each* company and plot them. (Marks = 2)** <br> [link text](https://)

```python
# Assuming df is your DataFrame and 'company' and 'price' are the column names
top_10_cars = df.groupby('company')['price'].nlargest(10)

print(top_10_cars)
```

```python
# first way
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the GroupBy object to a DataFrame and reset the index
top_10_cars_df = top_10_cars.to_frame().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='company', y='price', data=top_10_cars_df)
plt.title('Top 10 Highest Price Cars of Each Company')
plt.xticks(rotation=90)
plt.show()
```

```python
# second way
import matplotlib.pyplot as plt

# Group the dataframe by company and get the top 10 highest price cars for each company
top_10_cars = df.groupby('company').apply(lambda x: x.nlargest(10, 'price'))

plt.figure(figsize=(12, 10))
# bar horizontal
plt.barh(top_10_cars['company'] + ' - ' + top_10_cars['index'].astype(str), top_10_cars['price'])
plt.xlabel('Price')
plt.ylabel('Car')
plt.title('Top 10 Highest Price Cars of Each Company')
plt.show()
```
