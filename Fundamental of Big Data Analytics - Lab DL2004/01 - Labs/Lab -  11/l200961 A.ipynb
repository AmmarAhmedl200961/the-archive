```python
import string

# Sample input text
text = "Ali works on Sunday!"

# Mapper function to remove punctuation and convert to lower case
def mapper(text):
    for char in string.punctuation:
        text = text.replace(char, "")
    return text.lower()

# Reducer function to combine the mapped results
def reducer(mapped_results):
    return " ".join(mapped_results)

# Apply the mapper function to the input text
mapped_text = mapper(text)

# Apply the reducer function to the mapped text
result = reducer([mapped_text])

print(result)
```

```python
# Sample input data
data = ["Ali", "Ali", "John", "Ali"]

# Mapper function to count occurrences of each person
def mapper(data):
    counts = {}
    for person in data:
        if person in counts:
            counts[person] += 1
        else:
            counts[person] = 1
    return counts

# Reducer function to combine the mapped results
def reducer(mapped_results):
    combined_counts = {}
    for counts in mapped_results:
        for person, count in counts.items():
            if person in combined_counts:
                combined_counts[person] += count
            else:
                combined_counts[person] = count
    return combined_counts

# Apply the mapper function to the input data
mapped_counts = mapper(data)

# Apply the reducer function to the mapped counts
result = reducer([mapped_counts])

print(result)
```

```python
# Sample input data
data = {"Ali": ["Thursday", "Wednesday", "Wednesday"]}

# Mapper function to count occurrences of each person on each day
def mapper(data):
    day_counts = {}
    for person, days in data.items():
        counts = {}
        for day in days:
            if day in counts:
                counts[day] += 1
            else:
                counts[day] = 1
        day_counts[person] = counts
    return day_counts

# Reducer function to combine the mapped results
def reducer(mapped_results):
    combined_counts = {}
    for counts in mapped_results:
        for person, day_counts in counts.items():
            if person in combined_counts:
                for day, count in day_counts.items():
                    if day in combined_counts[person]:
                        combined_counts[person][day] += count
                    else:
                        combined_counts[person][day] = count
            else:
                combined_counts[person] = day_counts
    return combined_counts

# Apply the mapper function to the input data
mapped_day_counts = mapper(data)

# Apply the reducer function to the mapped day counts
result = reducer([mapped_day_counts])

print(result)
```

```python
# Sample input data
task2_data = {"Ali": 3, "John": 1}
task3_data = {'Ali': {'Thursday': 1, 'Wednesday': 2}}

# Mapper function to combine data from task 2 and task 3
def mapper(task2_data, task3_data):
    combined_data = {}
    for person, total_count in task2_data.items():
        if person in task3_data:
            day_counts = task3_data[person]
            for day, count in day_counts.items():
                combined_data[(person, day)] = (count, total_count)
    return combined_data

# Reducer function to combine the mapped results
def reducer(mapped_results):
    combined_counts = {}
    for person_day, counts in mapped_results.items():
        person, day = person_day
        if person in combined_counts:
            combined_counts[person].append((day, *counts))
        else:
            combined_counts[person] = [(day, *counts)]
    return combined_counts

# Apply the mapper function to the input data
mapped_combined_data = mapper(task2_data, task3_data)

# Apply the reducer function to the mapped combined data
result = reducer(mapped_combined_data)

print(result)
```
