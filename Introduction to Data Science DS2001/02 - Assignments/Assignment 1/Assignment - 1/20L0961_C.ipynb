###1.1 Write a function that reurns Count number of lines in txt file

```python
def NumOfLines(fileName):
  try:
    count = 0
    with open(fileName) as f:
      for line in f:
        count += 1  
    return count
  except FileNotFoundError:
    print("File not found")
```

```python
fileName = 'Text_File.txt'
count = NumOfLines(fileName)
print("Number of Lines are ", count)
```

###1.2 Write a function that reurns Count number of words in txt file

```python

def NumOfWords(fileName):
  try:
    words = 0
    with open(fileName) as f:
      for line in f:
        word = line.lower().split()
        # check if word is non numeric
        for part in word:
          if part.isalpha():
            words += 1
            
        
    return words
  except FileNotFoundError:
    print("File not found")
```

```python
num = NumOfWords(fileName)
print("Number of words are ", num)
```

##Task 2
Write a function that reurns a dataframe with two columns having stop words in one col and their count in the other.

```python
import pandas as pd

def CreateDF(fileName):
    try:
        # list of stopwords
        stopwords = ['and','or','is','the','a','an','to','of','it','as','in','on','that','this','for','with','at','be','are','by',\
            'from','was','were','has','have','had','can','could','would','should','will','shall','may','might','do','does','did','but','not']

        # read file content
        with open(fileName) as f:
            content = f.read().lower().split()

        # count occurrences of each word
        word_counts = {}
        for word in content:
            if word in stopwords:
                # check if word is in dictionary or not
                if word not in word_counts:
                    word_counts[word] = 1
                else:
                    word_counts[word] += 1

        # create DataFrame
        df = pd.DataFrame(list(word_counts.items()), columns=['Stopword', 'Count'], index=None)

        return df
    except:
        print("FILE NOT FOUND")
```

##Task 3
Now generate a txt file named RollnumberWS.txt having all content of the given file but excluding stopwords.

```python
def WithoutSW(fileName, filetowritein):
    stopwords = ['and','or','is','the','a','an','to','of','it','as','in','on','that','this','for','with','at','be','are','by',\
        'from','was','were','has','have','had','can','could','would','should','will','shall','may','might','do','does',\
        'did', 'but', 'not']

    try:
        with open(fileName, 'r') as f:
            content = f.read().lower().split()

        # remove stopwords
        content_without_sw = [word for word in content if word not in stopwords]

        # write to new file
        with open(filetowritein, 'w') as f:
            f.write(' '.join(content_without_sw))
    except FileNotFoundError:
        print(f'File {fileName} not found')
```

##Task 4
Write a function that displays all words having length greater than 5.

```python
def GreaterThan5(fileName):
    try:
        # read file content
        with open(fileName) as f:
            # skip \ufeff
            
            content = f.read().lower().split()

        # count occurrences of each word
        word_counts = {}
        for word in content:
            # edge case: check if word is non numeric
            if len(word) > 5 and word.isalpha():
                # check if word is in dictionary or not
                if word not in word_counts:
                    word_counts[word] = 1
                else:
                    word_counts[word] += 1


        return pd.DataFrame(list(word_counts.items()), columns=['Word', 'Count'], index=None)
    except:
        print("File not found")
```

##Task 5
Write a function that returns the count of all words which end at y.

```python
def EndAty(fileName):
    try:
        # read file content
        with open(fileName) as f:
            content = f.read().lower().split()
            
        # count occurrences of each word
        yEnding = {}
        for word in content:
            # edge case: check if word is non numeric
            if word.endswith('y') and word.isalpha():
                # check if word is in dictionary or not
                if word not in yEnding:
                    yEnding[word] = 1
                else:
                    yEnding[word] += 1

        yEnding = pd.DataFrame(list(yEnding.items()), columns=['Word', 'Count'], index=None)
        return yEnding
    except:
        print("File not found")
```

```python
def CountLower(fileName):
    try:
        with open(fileName) as f:
            content = f.read().split()
            
        # count occurrences of each word
        lower = {}
        for word in content:
            # edge case: check if word is non numeric
            if word.islower() and word.isalpha():
                # check if word is in dictionary or not
                if word not in lower:
                    lower[word] = 1
                else:
                    lower[word] += 1
                    
        return pd.DataFrame(list(lower.items()), columns=['Word', 'Count'], index=None)
    except:
        print("File not found")
```

##Task 7
Write a function to replace all "the" with "thee".

```python
def ReplaceThe(fileName):
    try:
        with open(fileName, 'r') as file:
            content = file.read()

        # Replace "the" with "thee" using the replace() method
        # further replace " the " with " thee " to avoid replacing "the" in other words
        # note: 'The' is not replaced
        content = content.replace(' the ', ' thee ')
        content = content.replace('the ', 'thee ')
        content = content.replace(' the', ' thee') 

        # Open the file in write mode and overwrite the content
        with open(fileName, 'w') as file:
            file.write(content)

        print(f'Check "{fileName}".')

    except FileNotFoundError:
        print(f'File "{fileName}" not found.')
```

##Task 8
Write a function ZICount() that prints the count of all occurences of Z and I (including small cases a and i too)
For example: Hi! I am Zenia. It was so good to see you!
####Output:
I or i : 4

Z or z : 1


```python
def ZICount(fileName):
    try:
        d = {'I or i': 0, 'Z or z': 0}
        with open(fileName) as f:
            # not using split as it has overhead of creating list
            content = f.read()
        for word in content:
            # get count of Z or z, and I or i, return dict{'I or i': count, 'Z or z': count}
            d['Z or z'] += word.count('z') + word.count('Z')
            d['I or i'] += word.count('i') + word.count('I')
        return d
    except FileNotFoundError:
        print(f'File "{fileName}" not found.')
```

##Task 9
Write a function to display the content of a file in ascending order (based on the word count)

Content: I am confused to start it but it won't be done until I give it a try

Output: I it to a am be confused done give start try until wont.


```python
def CountAsc(fileName):
    try:
        with open(fileName, 'r') as file:
            # Read the file and split it into words
            content_or_words = file.read().split()
        # Sort the words based on their length (word count)
        content_or_words.sort(key=lambda x: len(x))
        # Initialize an empty list to hold the final words
        final_words = []
        for word in content_or_words:
            # Check if the word is alphanumeric and not numeric
            if word.isalnum() and not word.isnumeric():
                # If the word is not already in the final words list, add it
                if word not in final_words:
                    final_words.append(word)
        print(final_words)
    except FileNotFoundError:
        print(f'File "{fileName}" not found.')
```

```python
CountAsc(fileName)
```

##Task 10
10.	Write a function alter_Upper that prints the content in a way that the first line is in Upper case, second in lower case and so on.

Example:
I am Zenia
I love painting
I live in Pakistan

Output:
i am zenia
I LOVE PAINTING
i live in pakistan


```python
def alter_Upper(fileName):
    try:
        with open(fileName, 'r') as file:
            lines = file.readlines()
        for i in range(len(lines)):
            if i % 2 == 0:
                print(lines[i].upper())
            else:
                print(lines[i].lower())
    except FileNotFoundError:
        print(f'File "{fileName}" not found.')
```
