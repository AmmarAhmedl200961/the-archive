# **FINDING OUTLIERS (NOISY DATA)**

```python
#PAGE 7
#4

def removing_outliers(d):
     Q1=d.quantile(0.25)
     Q3=d.quantile(0.75)
     IQR=Q3-Q1
     lowerlimit=Q1-1.5*IQR
     upperlimit=Q3+1.5*IQR
     return d[(d>lowerlimit)&(d<upperlimit)]


Age=removing_outliers(Age)
AMV1=removing_outliers(AMV)
Met=removing_outliers(Met)
Clo=removing_outliers(Clo)
Dewpt=removing_outliers(Dewpt)
PlaneRadTemp=removing_outliers(PlaneRadTemp)
Ta=removing_outliers(Ta)
Tmrt=removing_outliers(Tmrt)
Vel=removing_outliers(Vel)
AirTurb=removing_outliers(AirTurb)
Pa=removing_outliers(Pa)
Rh=removing_outliers(Rh)
PMV=removing_outliers(PMV)
TaOutdoor=removing_outliers(TaOutdoor)
RhOutdoor=removing_outliers(RhOutdoor)
```

# **ENTROPY FOLLOWED BY INFORMATION GAIN**

```python
#PAGE 10
def calculate_entropy(df):
    prob = df.value_counts(normalize = True)
    entropy = -1 * np.sum(np.log2(prob) * prob)
    return round(entropy, 3)
print('Entropy of Age Column: {}'.format(calculate_entropy(dataset['Age'])))   
print('Entropy of AMV column: {}'.format(calculate_entropy(dataset['AMV'])))
print('NEntropy of Met column: {}'.format(calculate_entropy(dataset['Met'])))
print('Entropy of Clo column: {}'.format(calculate_entropy(dataset['Clo'])))
print('Entropy of Dewpt column: {}'.format(calculate_entropy(dataset['Dewpt'])))
print('Entropy of PlaneRadTemp column: {}'.format(calculate_entropy(dataset['PlaneRadTemp'])))
print('Entropy of Ta column: {}'.format(calculate_entropy(dataset['Ta'])))
print('Entropy of Tmrt column: {}'.format(calculate_entropy(dataset['Tmrt'])))
print('Entropy of Vel column: {}'.format(calculate_entropy(dataset['Vel'])))
print('Entropy of AirTurb column: {}'.format(calculate_entropy(dataset['AirTurb'])))
print('Entropy of Pa column: {}'.format(calculate_entropy(dataset['Pa'])))
print('Entropy of Rh column: {}'.format(calculate_entropy(dataset['Rh'])))
print('Entropy of PMV column: {}'.format(calculate_entropy(dataset['PMV'])))
print('Entropy of TaOutdoor column: {}'.format(calculate_entropy(dataset['TaOutdoor'])))
print('Entropy of RhOutdoor column: {}'.format(calculate_entropy(dataset['RhOutdoor'])))

print(" ")
print("INFORMATION GAIN: ")
print(" ")

def calculate_IG(df, target, feature):
    target_entropy = calculate_entropy(df[target])
    entropy_list = []
    weight_list = []
    for i in df[feature].unique():
        feature_level = df[df[feature] == i]
        entropy_level = calculate_entropy(feature_level[target])
        entropy_list.append(round(entropy_level, 3))
        weight_level = len(feature_level) / len(df)
        weight_list.append(round(weight_level, 3))
    remaining_entropy = np.sum(np.array(entropy_list) * np.array(weight_list))
    information_gain = target_entropy - remaining_entropy
    return information_gain

dataset['AMV'] = dataset['AMV'].astype(int)
for i in dataset.drop(columns = 'AMV').columns:
    info_gain = calculate_IG(dataset, 'AMV', i)
    print(i)
    print(round(info_gain, 3))    
```

**FORWARD SELECTION FOR PMV**

```python
#PAGE 12
X=dataset.iloc[:,0]
X=dataset.iloc[:,2:12]
X['Age']=dataset.iloc[:,0]
X['TaOutdoor']=dataset.iloc[:,13:14]
X['RhOutdoor']=dataset.iloc[:,14:15]
Y=dataset.iloc[:,12:13]    #PMV
from sklearn.model_selection import train_test_split
X_train,Y_train,x_test,y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
from sklearn.ensemble import RandomForestClassifier

p_x = dataset[['Age', 'Vel', 'AirTurb', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
p_y = dataset['PMV']
#forward selection PMV linear regression
f_s = sfs(LinearRegression(), n_features_to_select = 5, direction = 'forward')
f_s.fit(p_x, p_y)
print(f_s.get_support())
X_train_pfs = X_train[['Age', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_train_pfs = y_train[['PMV']]
X_test_pfs = X_test[['Age', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_test_pfs = y_test[['PMV']]
lreg_ = LinearRegression()
lreg_.fit(X_train_pfs, y_train_pfs)
y_pred = lreg_.predict(X_test_pfs)
lreg_.score(X_test_pfs, y_pred)

                                                                    
```

**BACKWARD SELECTION OF PMV**

```python
#backward selection PMV linear regression
b_s = sfs(LinearRegression(), n_features_to_select = 5, direction = 'backward')
b_s.fit(p_x, p_y)
print(b_s.get_support())
X_train_pbs = X_train[['Age', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_train_pbs = y_train['PMV']
X_test_pbs = X_test[['Age', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_test_pbs = y_test[['PMV']]
lreg_ = LinearRegression()
lreg_.fit(X_train_pbs, y_train_pbs)
y_pred = lreg_.predict(X_test_pbs)
lreg_.score(X_test_pbs, y_pred)
```

**FORWARD SELECTION FOR AMV**

```python
#forward selection AMV logistic regression
a_x = dataset[['Age', 'Met', 'Clo', 'Vel', 'AirTurb', 'Pa', 'Rh', 'TaOutdoor', 'RhOutdoor']]
a_y = dataset['AMV']
f_s = sfs(LogisticRegression(), n_features_to_select = 5, direction = 'forward')
f_s.fit(a_x, a_y)
print(f_s.get_support())
X_train_afs = X_train[['Age', 'Vel','Rh', 'TaOutdoor', 'RhOutdoor']]
y_train_afs = y_train['AMV']
X_test_afs = X_test[['Age', 'Vel','Rh', 'TaOutdoor', 'RhOutdoor']]
y_test_afs = y_test['AMV']
logreg_ = LogisticRegression()
logreg_.fit(X_train_afs, y_train_afs)
y_pred = logreg_.predict(X_test_afs)
accuracy_score(y_test_afs, y_pred)
```

**BACKWARD SELECTION FOR AMV**

```python
#backward selection AMV logistic regression
f_s = sfs(LogisticRegression(), n_features_to_select = 5, direction = 'backward')
f_s.fit(a_x, a_y)
f_s.get_support()
X_train_abs = X_train[['Met', 'AirTurb', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_train_abs = y_train['AMV']
X_test_abs = X_test[['Met', 'AirTurb', 'Rh', 'TaOutdoor', 'RhOutdoor']]
y_test_abs = y_test['AMV']
logreg_ = LogisticRegression()
logreg_.fit(X_train_abs, y_train_abs)
y_pred = logreg_.predict(X_test_abs)
accuracy_score(y_test_abs, y_pred)

```
