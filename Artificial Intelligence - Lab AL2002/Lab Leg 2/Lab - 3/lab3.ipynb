```python
# graph for lab
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E', 'F'],
    'C': ['G'],
    'D': [],
    'E': [],
    'F': ['H'],
    'G': ['I'],
    'H': [],
    'I': []
}
```

```python
# bfs
def bfs(graph, node):
    visited = []  # Create an empty list to store visited nodes
    queue = []  # Create an empty queue to store nodes to be visited
    visited.append(node)  # Mark the starting node as visited and add it to the visited list
    queue.append(node)  # Add the starting node to the queue
    
    while queue:  # Continue the loop until the queue is empty
        s = queue.pop(0)  # Remove the first node from the queue and assign it to 's'
        print(s, end=" ")  # Print the value of 's' (the current node) with a space
        
        for neighbour in graph[s]:  # Iterate over the neighbors of the current node
            if neighbour not in visited:  # Check if the neighbor has not been visited
                visited.append(neighbour)  # Mark the neighbor as visited and add it to the visited list
                queue.append(neighbour)  # Add the neighbor to the queue for further exploration
                
bfs(graph, 'A')
```

```python
# dfs
def dfs(graph, node):
    visited = []  # Create an empty list to store visited nodes
    stack = [node]  # Create a stack and initialize it with the starting node

    while stack:  # Continue the loop until the stack is empty
        node = stack.pop()  # Pop the top node from the stack
        if node not in visited:  # Check if the node has not been visited
            visited.append(node)  # Add the node to the visited list
            stack.extend(reversed(graph[node]))  # Add the unvisited neighbors of the node to the stack in reverse order
            
    return visited  # Return the list of visited nodes

# Call the function
dfs(graph, 'A')
```
