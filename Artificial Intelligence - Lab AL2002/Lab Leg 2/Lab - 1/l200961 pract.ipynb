# Question 10

Create a function that takes a base number (b) and an exponent number (e) and returns the b^e calculation, without using built-in power function. Call the function in the next cell and show some outputs.
For Example:
myPower(2,3) return 8



```python
def power(b,e):
  mul=1
  for i in range(e):
    mul*=b
  return mul

print(power(2,3))
```

# Question 11

Write a function “Counter” that receives a character array and calculates Total characters, uppercases, lowercases and spaces then return them all. (In this question, all characters except spaces are counted as letters).

Sample Run:

Input:

Hello, I am a student of FAST University.

Values returned:

No. of letters are: 34

No. of spaces are: 7

No. of uppercase letters are: 7

No. of lowercase letters: 25





```python
def Counter(char_array):
    letters = 0
    uppercases = 0
    lowercases = 0
    spaces = 0
    for i in range(len(char_array)):
        if char_array[i].isalpha() or char_array[i]!= ' ':
            letters += 1
        if char_array[i].isupper():
            uppercases += 1
        elif char_array[i].islower():
            lowercases += 1
        elif char_array[i] == ' ':
            spaces += 1
    return letters, uppercases, lowercases, spaces

text="Hello, I am a student of FAST University."
letters,uppercase,lowercase,spaces = Counter(text)
print('No. of letters are:',letters)
print('No. of spaces are:',spaces)
print('No. of uppercase letters are:',uppercase)
print('No. of lowercase letters:',lowercase)
```
