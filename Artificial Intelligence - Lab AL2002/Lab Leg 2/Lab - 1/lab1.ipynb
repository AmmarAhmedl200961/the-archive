```python
def power(b,e):
  mul=1
  for i in range(e):
    mul*=b
  return mul

print(power(2,3))
```

```python
def Counter(char_array):
    letters = 0
    uppercases = 0
    lowercases = 0
    spaces = 0
    for i in range(len(char_array)):
        if char_array[i].isalpha() or char_array[i]!= ' ':
            letters += 1
        if char_array[i].isupper():
            uppercases += 1
        elif char_array[i].islower():
            lowercases += 1
        elif char_array[i] == ' ':
            spaces += 1
    return letters, uppercases, lowercases, spaces

text="Hello, I am a student of FAST University."
letters,uppercase,lowercase,spaces = Counter(text)
print('No. of letters are:',letters)
print('No. of spaces are:',spaces)
print('No. of uppercase letters are:',uppercase)
print('No. of lowercase letters:',lowercase)
```

```python
# Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of
# the two sorted arrays.
# The overall run time complexity should be O(log (m+n)).
# Example 1:
# Input: nums1 = [1,3], nums2 = [2]
# Output: 2.00000
# Explanation: merged array = [1,2,3] and median is 2.
# Example 2:
# Input: nums1 = [1,2], nums2 = [3,4]
# Output: 2.50000
# Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.

def findMedianSortedArrays(nums1, nums2):
    # add to nums1 the nums2
    nums1.extend(nums2)
    # sort the array
    nums1.sort()
    # if even then return the average of middle two elements
    if len(nums1)%2==0:
        return (nums1[int(len(nums1)/2)]+nums1[int(len(nums1)/2)-1])/2
    # else return the middle element
    else:
        return nums1[int(len(nums1)/2)]
    # time complexity is O(nlogn) because of sorting
    

print(findMedianSortedArrays([1,3],[2]))
print(findMedianSortedArrays([1,2],[3,4]))
```
