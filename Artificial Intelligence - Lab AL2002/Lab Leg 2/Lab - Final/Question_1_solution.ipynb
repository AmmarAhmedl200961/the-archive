```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

test_data = pd.read_excel("/content/ads and sales regression testing.xlsx")
train_data = pd.read_excel("/content/ads and sales regression training.xlsx")

model_regression = LinearRegression()

target_column1 = np.array(train_data["Ads"]).reshape(-1, 1)
pred_column1 = np.array(train_data["Sales"]).reshape(-1, 1)
target_column2 = np.array(test_data["ads"]).reshape(-1, 1)


model_regression.fit(target_column1, pred_column1)

pred_column2 = model_regression.predict(target_column2)

coefficient = model_regression.coef_
intercept = model_regression.intercept_
print("Predicted Value ", pred_column2)
print("coefficient ", coefficient)
print("intercept ", intercept)
```
