```python
# Question 3

# base class employee
class Employee:
    id = 0
    name = ""
    salary = 0.0

    def __init__(self, name: str, salary: float):
        # auto increment
        Employee.id += 1
        self.id = Employee.id
        self.name = name
        self.salary = salary

    def calculate_annual_salary(self):
        """a method that calculates annual salary based on the monthly salary"""
        return self.salary * 12

    def display_employee_details(self):
        print('Employee id: ', self.id)
        print('Employee Name: ', self.name)
        print('Employee salary: ', self.salary)


class Manager(Employee):
    bonus = 0.0  # bonus amount added to the monthly salary

    def __init__(self, name: str, salary: float, bonus: float):
        super().__init__(name, salary)
        Manager.bonus = bonus

    def calculate_annual_salary(self):
        return (self.salary + Manager.bonus) * 12


class Developer(Employee):
    programming_language = ""  # the primary programming language the developer works on

    def __init__(self, name: str, salary: float, programming_language: str):
        super().__init__(name, salary)
        Developer.programming_language = programming_language

    def display_employee_details(self):
        # call the parent class
        super().display_employee_details()
        print('Employee\'s programming language: ', Developer.programming_language)

# stack to store objects
employees = []


def add_to_both_emp(emp_name: str, monthly_salary: float, bonus=0.0, programming_language=""):
    """add new employees of both types"""
    if bonus == 0.0:
        emp = Developer(emp_name, monthly_salary, programming_language)
    else:
        emp = Manager(emp_name, monthly_salary, bonus)
    employees.append(emp)


def display_all_employees():
    """display all employees present in system"""
    for emp in employees:
        emp.display_employee_details()


# create instances of both developer and manager, add them to system and demonstrate use of function overloading and overriding
add_to_both_emp("John", 5000.0)
add_to_both_emp("Jane", 6000.0, bonus=1000.0)
add_to_both_emp("Bob", 7000.0, programming_language="Python")
display_all_employees()
```
