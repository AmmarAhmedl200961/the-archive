```python
import random
import math

GRID_SIZE = 20
POPULATION_SIZE = 5
GENERATIONS = 50
MUTATION_RATE = 0.1
PENALTY = 10000
# mines present in the grid
MINES = [(1,20),(5,20),(17,20),(11,19),(3,16),(3,14),(3,14),(16,13),(5,12),(2,10),(20,10),(17,7),(1,6),(10,6),(20,4),(4,3),(19,2),(16,1)]

# Define a function to generate a chromosome with fixed start and end points
def generate_chromosome():
    chromosome = [(0, 0) for _ in range(5)]  # initialize the chromosome with (0, 0) coordinates
    chromosome[0] = (0,0)  # set the start point 
    chromosome[-1] = (19,19) # set the end point 
    for i in range(1, 4):  # set the intermediate points to random locations
        chromosome[i] = (random.randint(0, GRID_SIZE-1), random.randint(0, GRID_SIZE-1))
    return chromosome


# Define a function to calculate the fitness of a chromosome  
def calculate_fitness(chromosome):
    distance = 0
    prev_x, prev_y = chromosome[0]
    mines_count = 0
    for i in range(1, len(chromosome)):
        x, y = chromosome[i]
        distance += math.sqrt((x - prev_x) ** 2 + (y - prev_y) ** 2)
        if (x, y) in MINES:
            mines_count += 1
        prev_x, prev_y = x, y
    fitness = distance + mines_count * PENALTY
    return fitness

# Define a function to select parents for crossover 
def selection(population):
    fitnesses = [calculate_fitness(chromosome) for chromosome in population]
    fitness_range = max(fitnesses) - min(fitnesses)
    probabilities = [(max(fitnesses) - fitness) / fitness_range for fitness in fitnesses]
    parents = random.choices(population, weights=probabilities, k=2)
    return parents

# Define a function to perform crossover between two chromosomes
def crossover(parent1, parent2):
    child1 = [(0, 0) for _ in range(5)]
    child2 = [(0, 0) for _ in range(5)]
    
    # Two-Point Crossover
    p1, p2 = sorted(random.sample(range(1, len(parent1)), 2))
    # swap the genes between the two points for the two children
    child1[p1:p2], child2[p1:p2] = parent2[p1:p2], parent1[p1:p2]
    
    # Uniform Crossover
    for i in range(len(parent1)):
        if parent1[i] not in child1:
            child1[i] = parent1[i] if random.random() < 0.5 else parent2[i]
        if parent2[i] not in child2:
            child2[i] = parent2[i] if random.random() < 0.5 else parent1[i]
    
    return child1, child2
    
# Define a function to perform mutation on a chromosome
def mutate(chromosome):
# Initialize the population with random chromosomes
    for i in range(len(chromosome)):
        if random.random() < MUTATION_RATE:
            chromosome[i] = (random.randint(0, GRID_SIZE-1), random.randint(0, GRID_SIZE-1))
    return chromosome    

population = [generate_chromosome() for _ in range(POPULATION_SIZE)]
print("pop: ",population, "\n")

# Evolve the population for a fixed number of generations
for generation in range(GENERATIONS):
    # Select parents for crossover
    parents = selection(population)
    
    # Perform crossover to generate a child
    child1, child2 = crossover(parents[0], parents[1])
    
    # Mutate the child
    child1 = mutate(child1)
    child2 = mutate(child2)
    
    # Replace a least fit chromosome in the population with the child
    fitnesses = [calculate_fitness(chromosome) for chromosome in population]
    min_fitness_index = fitnesses.index(min(fitnesses))
    if calculate_fitness(child1) > calculate_fitness(child2):
        population[min_fitness_index] = child2
    else:
        population[min_fitness_index] = child1
    
    # Print the fittest chromosome in the population
    fittest_chromosome = min(population, key=calculate_fitness)
    print(f"Generation {generation}: Fittest chromosome - {fittest_chromosome}, Fitness - {calculate_fitness(fittest_chromosome)}")
```

```python
import numpy as np

# Define the input data for the AND gate
input_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Define the target output for the AND gate
target_output = np.array([[0], [0], [0], [1]])

# Define the learning rate and the number of iterations
learning_rate = 0.1
num_iterations = 10

# Define the activation function (step function)
def activation_function(x):
    return 1 if x >= 0 else 0

# Initialize the weights and the bias
weights = np.array([0.0, 0.0])
bias = 0.0

# Train the perceptron
for i in range(num_iterations):
    for j in range(len(input_data)):
        x = input_data[j]
        y = target_output[j]
        # Calculate the weighted sum of the inputs and the bias
        weighted_sum = np.dot(x, weights) + bias
        # Calculate the predicted output using the activation function
        predicted_output = activation_function(weighted_sum)
        # Calculate the error
        error = y - predicted_output
        # Update the weights and the bias
        weights += learning_rate * error * x
        bias += learning_rate * error * 1
        
# Print the learned weight and bias
print("Learned weights:", weights)
print("Learned bias:", bias)

# Test the perceptron
test_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
print("Input\tPredicted Output")
for i in range(len(test_inputs)):
    x = test_inputs[i]
    weighted_sum = np.dot(x, weights) + bias
    predicted_output = activation_function(weighted_sum)
    # now print the input and the predicted output like a logical table
    print(f"{x}\t{predicted_output}")
```
