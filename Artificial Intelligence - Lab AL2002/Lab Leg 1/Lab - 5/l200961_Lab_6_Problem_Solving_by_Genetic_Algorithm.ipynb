Let say we have fucntion that evaluates the expression to 0 if the answer of expression " 6 * x ** 3 + 9 * y ** 2 + 90 * z " is 25. 
So That's mean we need the most suitable values of x, y and z so that we could achive our target value whichis 25 in this case. 

```python
def evaluateExpression(x, y, z):
	return 6 * x ** 3 + 9 * y ** 2 + 90 * z - 25
```

Let's write the solution based on Genetic Algorithm technique to find the most suitable values of x, y and z. 

*Step1* : Population of Solutions: 
Population is generated entirely from random numbers let say upto 1000 indivuduals. 

```python
# The algorithm starts by creating a population of 1000 solutions, with each solution consisting of three random values of x, y, and z.
# generate solutions
solutions = []
for counter in range(1000):
	solutions.append((random.uniform(0, 1000), random.uniform(0, 1000), random.uniform(0, 1000)))
```

Step2: Fitness fucntion: 
So the most fittest solution will be the one which evalutes the expression to "0". Otherwise the best solution will be closest to zero. 
So the fittness in this case can be seen if the solution is closest to zero. Hnece we can return highest fitness value to those solution which are closest to zero. 

```python
# Next, a fitness function is defined that takes in the values of x, y, and z for a given solution, 
# evaluates the expression using the evaluateExpression() function, and returns a fitness score. 
# The fitness score is the absolute value of 1 divided by the difference between the evaluated expression and the target value of 25. 
# The higher the fitness score, the better the solution.
def fitness(x, y, z):
	ans = evaluateExpression(x, y, z)

	if ans == 0:
		return 99999
	else:
		return abs(1 / ans)
```

Step3: Mating, Crossover or Generating the Genrations: 
During each generations further sub steps are performed like : 
Step 3.1: Selection of top ranked solutions 
Step 3.2: Mutation or slight changes or variation in values of solution. 
P.S(Here for the sake of anology if solution can be considered as chromosome then vatiable values can be considered as genes)
 

```python
# The algorithm then proceeds to perform 10000 generations of evolution. During each generation, 
# the solutions in the population are ranked based on their fitness scores, 
# and the top 100 solutions are selected as the "parents" for the next generation. 
# The values of x, y, and z for these top solutions are used to create a pool of "genes" that will be used for mutation in the next step.
for generation_count in range(10000):
	rankedSolutions = []
  # fitness step
	for solution in solutions:
		rankedSolutions.append((fitness(solution[0], solution[1], solution[2]), solution))
	rankedSolutions.sort()
	rankedSolutions.reverse()
	print(f"=== Generation {generation_count} best solutions ====")
	print(rankedSolutions[0])

	if rankedSolutions[0][0] > 999:
		break

	bestSolution = rankedSolutions[:100]
	# print(bestSolution)

  # selection step
	variables = []
	for solution in bestSolution:
		variables.append(solution[1][0])  # variable x
		variables.append(solution[1][1])  # variable y
		variables.append(solution[1][2])  # variable z

	newGeneration = []
  # mutation step
  # To create the next generation of solutions, the algorithm selects a random gene from the pool, 
  # and mutates its value by multiplying it by a random number between 0.99 and 1.01. 
  # This introduces a slight variation in the value of the gene, which helps to explore new regions of the search space.
	for counter in range(1000):
		x = random.choice(variables) * random.uniform(0.99, 1.01)
		y = random.choice(variables) * random.uniform(0.99, 1.01)
		z = random.choice(variables) * random.uniform(0.99, 1.01)

		newGeneration.append((x, y, z))
	# Finally, the new solutions are added to the population, and the process repeats for the next generation
	solutions = newGeneration
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

# Define fitness function
def fitness_function(individual):
    
    # Select features haveing 1 value 
    selected_features = [i for i in range(len(individual)) if individual[i] == 1]
    
    """
     # Train classifier and return accuracy based on given chromosome
    """
    X_train_selected = X_train.iloc[:,selected_features]
    X_test_selected = X_test.iloc[:, selected_features]
    dtc = DecisionTreeClassifier(random_state=42)
    dtc.fit(X_train_selected, y_train)
    y_pred = dtc.predict(X_test_selected)
    return accuracy_score(y_test, y_pred)
```

```python
# Define genetic algorithm parameters
population_size = 100
generations = 50
```

```python
# Define initial population
def create_individual():
    return [random.randint(0, 1) for _ in range(len(X_train.columns))]
    
```

```python
def create_population():
    return [create_individual() for _ in range(population_size)]
```

```python
# Define genetic operators
def crossover(parent1, parent2):
    # Choose crossover point
    crossover_point = random.randint(0, len(parent1) - 1)
    # Swap tails of parents
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def mutation(individual, probability=0.05):
    # Mutate individual
    for i in range(len(individual)):
        if random.random() < probability:
            individual[i] = 1 if individual[i] == 0 else 0
    return individual
```

```python
population = create_population()
print(population)
```

```python
# Run genetic algorithm

for generation in range(generations):
    # Select two parents based on fitness 
    fitness_scores = [fitness_function(individual) for individual in population]
    best_individual_idx = max(range(population_size), key=lambda i: fitness_scores[i])
    best_individual = population[best_individual_idx]
    print(f'Generation {generation+1}, Best fitness: {fitness_scores[best_individual_idx]}')
    next_generation = [best_individual]
    # Create two children child1 and child2 by crossover
    while len(next_generation) < population_size:
        parent1_idx = random.randint(0, population_size - 1)
        parent2_idx = random.randint(0, population_size - 1)
        parent1 = population[parent1_idx]
        parent2 = population[parent2_idx]
        child1, child2 = crossover(parent1, parent2)
        # population = next_generation

        # Mutate children
        child1 = mutation(child1, probability=0.05)
        child2 = mutation(child2, probability=0.05)
        next_generation.extend([child1, child2])
        

    # Evaluate fitness of children
    fitness_scores = [fitness_function(individual) for individual in population]
    best_individual_idx = max(range(population_size), key=lambda i: fitness_scores[i])
    best_individual = population[best_individual_idx]
    selected_features = [i for i in range(len(best_individual)) if best_individual[i] == 1] 

    # Replace worst individual with better child 
    worst_individual_idx = min(range(population_size), key=lambda i: fitness_scores[i])
    if fitness_scores[best_individual_idx] > fitness_scores[worst_individual_idx]:
        population[worst_individual_idx] = best_individual
        
    # Check if the best individual meets the termination condition
    if fitness_scores[best_individual_idx] > 0.9:
        break
        
    # Replace the current population with the next generation
    population = next_generation
    # pass
    
    
    

# Select best individual based on fitness
fitness_scores = [fitness_function(individual) for individual in population]
best_individual_idx = max(range(population_size), key=lambda i: fitness_scores[i])
best_individual = population[best_individual_idx]

# Print selected features
selected_features = [i for i in range(len(best_individual)) if best_individual[i] == 1] 
print(f'Selected Features: {selected_features}')

# the cross validation score of the best individual 
X_train_selected = X_train.iloc[:,selected_features]
dtc=DecisionTreeClassifier(random_state=42)
print(f'Cross Validation Score: {cross_val_score(dtc, X_train_selected, y_train, cv=5).mean()}')
```
