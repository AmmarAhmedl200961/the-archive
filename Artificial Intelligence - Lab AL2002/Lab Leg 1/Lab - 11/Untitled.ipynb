```python
numiter = 1000
# initialize variables
W14=0.2
W15=-0.3
W24 =0.4
W25 =0.1
W34 =-0.5
W35 =0.2
W46 =-0.3
W56 =-0.2
W4 =-0.4 
W5 =0.2 
W6=0.1

wlist=[W14,
W15,
W24,
W25,
W34,
W35,
W46,
W56,
W4 ,
W5 ,
W6]

b1=1
b2=1
blist=[b1,b2]

# input and target variables
x1=1
x2=0
x3=1
xlist=[x1,x2,x3]

t1=1
tlist=[t1]

# set learning rate 
alpha = 0.9
```

```python
def sigmoid(x):
    return np.divide(1,(1+np.exp(-x)))
def forwardprop(xlist,wlist,blist):
    # calculate hidden layer 1
    h1 = sigmoid(xlist[0]*wlist[0] + xlist[1]*wlist[2] + xlist[2]*wlist[4] + blist[0])
    # calculate hidden layer 2
    h2 = sigmoid(xlist[0]*wlist[1] + xlist[1]*wlist[3] + xlist[2]*wlist[5] + blist[0])

    # calculate output layer
    o1 = sigmoid(h1*wlist[6] + h2*wlist[7] + blist[1])

    return [h1, h2, o1]

def backpropagate(xlist, wlist, blist, tlist, alpha):
    # forward propagate to calculate outputs
    h1, h2, o1 = forwardprop(xlist, wlist, blist)
    
    # calculate error
    err = tlist[0] - o1
    
    # calculate derivative of error with respect to output
    dEdo1 = -2*err
    
    # calculate derivatives of error with respect to weights and biases
    dEdw6 = dEdo1 * o1 * (1 - o1) * h1
    dEdw7 = dEdo1 * o1 * (1 - o1) * h2
    dEdh1 = dEdo1 * o1 * (1 - o1) * w6
    dEdh2 = dEdo1 * o1 * (1 - o1) * w7
    dEdw4 = dEdh1 * h1 * (1 - h1) * xlist[0]
    dEdw5 = dEdh1 * h1 * (1 - h1) * xlist[2]
    dEdw8 = dEdh2 * h2 * (1 - h2) * xlist[0]
    dEdw9 = dEdh2 * h2 * (1 - h2) * xlist[2]
    dEdb1 = dEdh1 * h1 * (1 - h1) + dEdh2 * h2 * (1 - h2)
    dEdb2 = dEdo1 * o1 * (1 - o1)
    
    # update weights and biases
    w6 -= alpha * dEdw6
    w7 -= alpha * dEdw7
    w4 -= alpha * dEdw4
    w5 -= alpha * dEdw5
    w8 -= alpha * dEdw8
    w9 -= alpha * dEdw9
    b1 -= alpha * dEdb1
    b2 -= alpha * dEdb2
    
    # return updated weights
    return [w4, w5, w6, w7, w8, w9, b1, b2]

def error(olist,tlist):
    return 0.5 * (np.power(oList[0] - tList[0], 2) + np.power(oList[1] - tList[1], 2))
```
