# **4. Data Types**
R supports several data types including:

Numeric,
Integer,
Logical (TRUE/FALSE)
Character (String)

```r
num_var <- 3.14  # Numeric
int_var <- 5L    # Integer (denoted by the L suffix)
bool_var <- TRUE # Logical
char_var <- "R Language" # Character
comp <- 9 + 3i #complex

# Displaying data types
print(class(num_var))
print(class(int_var))
print(class(bool_var))
print(class(char_var))
print(class(comp))
```

```r
x <- 1L # integer
y <- 2 # numeric

# convert from integer to numeric:
a <- as.numeric(x)

# convert from numeric to integer:
b <- as.integer(y)

# print values of x and y
x
y

# print the class name of a and b
class(a)
class(b)
```

# **5. Operators**
Operators include arithmetic, comparison, and logical operators.

**Arithmetic Operators:**

```r
a <- 10
b <- 3
sum <- a + b
diff <- a - b
prod <- a * b
quotient <- a / b
rounded_quotient <- a %/% b
remainder <- a %% b
exp <- a ^ b

print(c(sum, diff, prod, quotient, rounded_quotient, remainder, exp))
```

**Comparison Operators:**

```r
a <- 10
b <- 20
print(a == b)  # Equal to
print(a != b)  # Not equal to
print(a > b)   # Greater than
print(a < b)   # Less than
```

**Logical Operators:**

```r
print(TRUE & FALSE)  # AND
print(TRUE | FALSE)  # OR
print(!TRUE)         # NOT
```

**Other Operators:**
*   **:**	Creates a series of numbers in a sequence	**x <- 1:10**
*   **%in%**	Find out if an element belongs to a vector	**x %in% y**
*   `%*%` 	Matrix Multiplication	x <- Matrix1 `%*% `Matrix2




# **6. Conditions**
You can use conditional statements like if, else if, and else.

```r
x <- 10

if (x > 0) {
  print("Positive number")
} else if (x == 0) {
  print("Zero")
} else {
  print("Negative number")
}
```

# **7. Loops**
R supports various loops including for, while, and repeat.
break and next statement

# **8. Functions**
Functions are used to encapsulate reusable code. You can define functions using the function keyword.

```r
# Function to calculate the square of a number
square <- function(x) {
  return(x * x)
}

result <- square(4)
print(result)
```

```r
#Global Variable

txt <- "awesome"
my_function <- function() {
  txt <<- "fantastic"
  paste("R is", txt)
}

my_function()

paste("R is", txt)
```

**2. Pie Charts**

By default, the plotting of the first pie starts from the x-axis and move counterclockwise.

```r
# Create a vector of pies
x <- c(10,20,30,40)

# Display the pie chart
pie(x)
```

```r
# Display the pie chart and start the first pie at 90 degrees
pie(x, init.angle = 90)
```

```r
# Create a vector of pies
x <- c(10,20,30,40)

# Create a vector of labels
mylabel <- c("Apples", "Bananas", "Cherries", "Dates")
# Create a vector of colors
colors <- c("blue", "yellow", "green", "black")

# Display the pie chart with labels
pie(x, label = mylabel, main = "Fruits", col = colors)
# Display the explanation box
legend("bottomright", mylabel, fill = colors)
```

The legend can be positioned as either:

bottomright, bottom, bottomleft, left, topleft, top, topright, right, center
