```python
rows = table.find_all('tr')
```

```python
headers = rows[0]
```

```python
headers
```

```python
head = headers.find_all('td')
```

```python
head[0].text.strip()
```

```python
titles = []
for i in head:
    title = i.text.strip()
    titles.append(title)
```

```python
titles
```

```python
df = pd.DataFrame(columns = titles)
```

```python
df
```

```python
for i in rows[1:]:
    data = i.find_all('td')
    row = [tr.text.strip() for tr in data]
    l = len(df)
    df.loc[l] = row
```

```python
content = driver.find_element(by=By.XPATH, value='/html/body/div[4]/div/div/div[2]/div[4]/div/div[7]/div[2]/main/section/div')
```

```python
html = BeautifulSoup(content.get_attribute("outerHTML"), 'html.parser')
```

```python
all_products = html.find_all("div", class_="product-card__info disable-animations")
```

```python
len(all_products)
```

```python
names = []
colors = []
prices=[]
for product in all_products:
    name = product.find("div", class_="product-card__title")
    color = product.find("div", class_="product-card__product-count")
    price = product.find("div", class_ = "product-card__price")
    names.append(name.text.strip())
    colors.append(color.text.strip())
    prices.append(price.text.strip())
```

```python
df2 = pd.DataFrame()
```

```python
df2["Product Name"] = names
df2["Color Options"] = colors
df2["Price"] = prices
```

```python
df2
```

```python
df2.to_csv("nike.csv")
```
