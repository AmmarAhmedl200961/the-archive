```python
# a psp scraper
from bs4 import BeautifulSoup
import requests

base_url = "https://www.pspunk.com/psp-themes/"
ptf_has_pagination = "https://www.pspunk.com/psp-ptf-themes/"
ctf_has_pagination = "https://www.pspunk.com/psp-ctf-themes/"

def scrape_psp_themes(base_url, ptf_has_pagination, ctf_has_pagination):
    # scrape psp themes
    for url in [base_url, ptf_has_pagination, ctf_has_pagination]:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        links = soup.find_all('a', attrs={'data-type': 'URL'})
        for link in links:
            print (link['href'])
        if url == ptf_has_pagination:
            pages = [i for i in range(0, 5+1)]
            for page in pages:
                response = requests.get(f"{url}/{page}")
                soup = BeautifulSoup(response.text, 'html.parser')
                links = soup.find_all('a')
                for link in links:
                    if '.ptf' in link['href']:
                        print(link['href'])
        if url == ctf_has_pagination:
            pages = [i for i in range(0, 18+1)]
            for page in pages:
                response = requests.get(f"{url}/{page}")
                soup = BeautifulSoup(response.text, 'html.parser')
                links = soup.find_all('a')
                for link in links:
                    if '.ctf' in link['href']:
                        print(link['href'])        
        
        

scrape_psp_themes(base_url, ptf_has_pagination, ctf_has_pagination)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
import time

urls = [
    "https://cdromance.org/psp/grand-theft-auto-vice-city-stories-ps2-assets-hack/",
    "https://cdromance.org/psp/grand-theft-auto-liberty-city-stories-ps2-assets-hack/",
    "https://cdromance.org/psx2psp-eboots/spider-man-usa/",
    "https://cdromance.org/psx2psp-eboots/silent-hill-uncensored-europe/",
    "https://cdromance.org/psx2psp-eboots/dino-crisis-usa/",
    "https://cdromance.org/psx2psp-eboots/dino-crisis-2-usa/",
    'https://cdromance.org/psx2psp-eboots/driver-2-usa-psx2psp/'
]

# Initialize the Chrome driver
driver = webdriver.Chrome()

for url in urls:
    # Load the webpage
    driver.get(url)
    
    try:
        # Wait until the button is clickable
        wait = WebDriverWait(driver, 10)
        # scroll to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "acf-get-content-button")))
        
        # Click the button
        button.click()
        
        # Wait for the content to load
        time.sleep(5)  # Adjust the sleep time as needed
        
        # Get the page source and parse it with BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Extract the links (example: all <a> tags)
        links = soup.find_all('a', id="dl-btn-0")
        for link in links:
            print(link.get('href'))
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Quit the driver
driver.quit()
```
