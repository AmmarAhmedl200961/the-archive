```python
import pandas as pd
df=pd.read_csv(r'Labelled  Tweets.csv')
print(df)
```

```python
from sklearn.model_selection import train_test_split
X=df['Tweets'].values
y=df['Class'].values
##split into test/train
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
##Bag of words(no of unique words)
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer()
bow_train=vect.fit_transform(X_train)
bow_test=vect.transform(X_test)
print(bow_test)
```

```python
#TF-IDF(term freq-inverse doc freq.)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect=TfidfVectorizer()
tfidf_train=tfidf_vect.fit_transform(X_train)
tfidf_test=vect.transform(X_test)
print(tfidf_test)

```

```python
#Preprocess using Inbuit libraries
##Tokenize
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
tokenize=[word_tokenize(X)for X in X]
for i, tokens in enumerate(tokenize):
    print(f"Tokenized Words for Tweet {i + 1}: {tokens}")
```

```python
##Lowecase Tokens
lowercaseTokens=[[token.lower()for token in tokens]for tokens in tokenize]
for i, tokens in enumerate(lowercaseTokens):
    print(f"Lowercase Words for Tweet {i + 1}: {tokens}")
```

```python
##Stop English words(remove common english)
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
filtered_tokens=[[token for token in tokens if token not in stop_words]for tokens in lowercaseTokens]
for i, tokens in enumerate(filtered_tokens):
    print(f"Filtered Words for Tweet {i + 1}: {tokens}")
```

```python
##LaplaceSmoothing

from collections import defaultdict
# Laplace smoothing for Bag of Words (BoW) with handling out-of-vocabulary words
class LaplaceSmoothingBoW:
    def __init__(self, alpha=1):
        self.alpha = alpha

    def fit(self, tokenized_sentences):
        counts = defaultdict(int)
        total_words = 0
        vocabulary = set()

        # Count occurrences of each word
        for tokens in tokenized_sentences:
            for token in tokens:
                counts[token] += 1
                total_words += 1
                vocabulary.add(token)

        # Special token for out-of-vocabulary words
        out_of_vocab_token = 'OOV'
        vocabulary.add(out_of_vocab_token)

        def get_probability(word):
            count = counts.get(word, 0)
            probability = (count + self.alpha) / (total_words + self.alpha * len(vocabulary))
            return probability

        return get_probability, out_of_vocab_token

# Apply Laplace smoothing to filtered_tokens
laplace_bow_model = LaplaceSmoothingBoW()
laplace_probability_function, out_of_vocab_token = laplace_bow_model.fit(filtered_tokens)

# Test Laplace smoothing on some words
test_words = ['the', 'cat', 'dog', 'unseenword']
laplace_probabilities = {word: laplace_probability_function(word) for word in test_words}

# Print Laplace smoothing probabilities
for word, probability in laplace_probabilities.items():
    print(f'Probability of "{word}" with Laplace smoothing: {probability}')

# Example for out-of-vocabulary word
oov_probability = laplace_probability_function(out_of_vocab_token)
print(f'Probability of "{out_of_vocab_token}" with Laplace smoothing: {oov_probability}')
```

```python
#Remove punctaution
import string
def remove_punctuation(text):
    return ''.join(char for char in text if char not in string.punctuation)
df['normalized_text'] = df['Tweets'].apply(remove_punctuation)
print(df['normalized_text'] )
```

```python
#Stemming(Base Form)
from nltk.stem import PorterStemmer
porter = PorterStemmer()
def apply_stemming(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Apply stemming to each word
    stemmed_words = [porter.stem(word) for word in words]

    # Join the stemmed words to form the normalized text
    normalized_text = ' '.join(stemmed_words)

    return normalized_text
df['stemmed_text'] = df['normalized_text'].apply(apply_stemming)

print(df['stemmed_text'][0])
```

```python
#Lemmization(Dictionary form)
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
def apply_lemmatization(text):
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    normalized_text = ' '.join(lemmatized_words)
    return normalized_text
df['lemmatized_text'] = df['normalized_text'].apply(apply_lemmatization)
print(df['lemmatized_text'][0])
```

```python
#Logistic Regression for classification
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Function to initialize weights and bias
def initialize_params(num_features):
    weights = np.zeros(num_features)
    bias = 0
    return weights, bias

# Function to compute the cost (binary cross-entropy)
def compute_cost(X, y, weights, bias):
    m = X.shape[0]
    y_pred = sigmoid(np.dot(X, weights) + bias)
    cost = -1/m * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
    return cost

# Function to perform gradient descent and update parameters
def gradient_descent(X, y, weights, bias, learning_rate, epochs):
    m = X.shape[0]

    for epoch in range(epochs):
        # Compute predicted probabilities
        y_pred = sigmoid(np.dot(X, weights) + bias)

        # Compute gradients
        dw = (1/m) * np.dot(X.T, (y_pred - y))
        db = (1/m) * np.sum(y_pred - y)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        # Compute and print the cost every 100 epochs
        if epoch % 100 == 0:
            cost = compute_cost(X, y, weights, bias)
            print(f"Epoch {epoch}, Cost: {cost}")

    return weights, bias

# Function to make predictions
def predict(X, weights, bias):
    return sigmoid(np.dot(X, weights) + bias)

# Function to evaluate accuracy
def accuracy(y_true, y_pred):
    return np.mean(y_true == (y_pred >= 0.5).astype(int))

# Take a few hundred rows for training and testing
df_subset = df.sample(n=300, random_state=42)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_subset['Tweets'], df_subset['Class'], test_size=0.20, random_state=42)

# Bag-of-Words representation
vectorizer_bow = CountVectorizer()
bow_train = vectorizer_bow.fit_transform(X_train)
bow_test = vectorizer_bow.transform(X_test)

# TF-IDF representation
vectorizer_tfidf = TfidfVectorizer()
tfidf_train = vectorizer_tfidf.fit_transform(X_train)
tfidf_test = vectorizer_tfidf.transform(X_test)

# Convert sparse matrices to NumPy arrays
bow_train = bow_train.toarray()
bow_test = bow_test.toarray()
tfidf_train = tfidf_train.toarray()
tfidf_test = tfidf_test.toarray()

# Initialize parameters for BOW
weights_bow, bias_bow = initialize_params(bow_train.shape[1])

# Train logistic regression model using BOW representation
weights_bow, bias_bow = gradient_descent(bow_train, y_train, weights_bow, bias_bow, learning_rate=0.01, epochs=500)

# Initialize parameters for TF-IDF
weights_tfidf, bias_tfidf = initialize_params(tfidf_train.shape[1])

# Train logistic regression model using TF-IDF representation
weights_tfidf, bias_tfidf = gradient_descent(tfidf_train, y_train, weights_tfidf, bias_tfidf, learning_rate=0.5, epochs=500)

# Make predictions on test data
y_pred_bow = predict(bow_test, weights_bow, bias_bow)
y_pred_tfidf = predict(tfidf_test, weights_tfidf, bias_tfidf)

# Evaluate accuracy
accuracy_bow = accuracy(y_test, y_pred_bow)
accuracy_tfidf = accuracy(y_test, y_pred_tfidf)

print("Accuracy using BOW:", accuracy_bow)
print("Accuracy using TF-IDF:", accuracy_tfidf)
```

```python
#Evaluate Performance
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='binary')
    recall = recall_score(y_true, y_pred, average='binary')
    f1 = f1_score(y_true, y_pred, average='binary')

    return accuracy, precision, recall, f1
y_pred_bow_binary = (y_pred_bow >= 0.5).astype(float)
accuracy_bow, precision_bow, recall_bow, f1_bow = evaluate_metrics(y_test, y_pred_bow_binary)


print("Metrics for BOW:")
print(f"Accuracy: {accuracy_bow}")
print(f"Precision: {precision_bow}")
print(f"Recall: {recall_bow}")
print(f"F1-Score: {f1_bow}")
print()
```

```python
y_pred_tfidf_binary = (y_pred_tfidf >= 0.5).astype(float)
accuracy_tfidf, precision_tfidf, recall_tfidf, f1_tfidf = evaluate_metrics(y_test, y_pred_tfidf_binary)
print("Metrics for TF-IDF:")
print(f"Accuracy: {accuracy_tfidf}")
print(f"Precision: {precision_tfidf}")
print(f"Recall: {recall_tfidf}")
print(f"F1-Score: {f1_tfidf}")
```
