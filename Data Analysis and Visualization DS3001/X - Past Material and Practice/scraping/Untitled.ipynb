```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time

# Set the path to your Chrome driver executable
driver_path = 'C:/Users/ahmad/Downloads/chromedriver-win64/chromedriver.exe'

# Initialize the Chrome service and driver
service = Service(executable_path=driver_path)
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service=service, options=options)

# URL to scrape
url = 'https://www.youtube.com/watch?v=qtj2OqmDT1I&ab_channel=EpidemicPop'

# Load the webpage and maximize the window
driver.get(url)
driver.maximize_window()
time.sleep(2)

# Scroll down to load more comments
for _ in range(2):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(3)

# Get top-level comments
comments = driver.find_elements(By.XPATH, '//*[@id="content-text"]')
arr = []

for comment in comments:
    arr.append(comment.text)

    # Get replies for each comment
    try:
        reply_button = comment.find_element(By.XPATH, './/[@id="expander"]')
        reply_button.click()
        time.sleep(2)

        # Get reply comments
        reply_comments = comment.find_elements(By.XPATH, './/[@id="content-text"]')
        for reply_comment in reply_comments:
            arr.append(reply_comment.text)
    except Exception as e:
        print(f"Error while fetching replies: {e}")

# Create a DataFrame for the comments and replies
df = pd.DataFrame({'Comments': arr})

# Save the DataFrame to a CSV file
df.to_csv('youtube_comments_single_video.csv', index=False)

# Close the WebDriver
driver.quit()
```
