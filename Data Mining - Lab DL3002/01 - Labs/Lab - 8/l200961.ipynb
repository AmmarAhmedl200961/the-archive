```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('groceries_dataset.csv')
```

```python
records = []
for i in range(0, 9835):
    records.append([str(df.values[i,u]) for u in range(0, 32)])
```

```python
rules = apriori(records, min_support=0.0022, min_confidence=0.20, min_lift=3, min_length = 4)
```

```python
# Store rules in result variable
results = list(rules)# See the items that were bought together with their support
results_list = []
for i in range(0, len(results)):
    results_list.append('RULE:\t' + str(results[i][0]) + '\nSUPPORT:\t' + str(results[i][1]))
```

```python
import pandas as pd
data = pd.read_csv("DATA.csv")
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race','sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
data.columns = columns

# Keep only the first 500 rows of the dataset
# data = data.head(500)
# data_processed = pd.get_dummies(data)
# data_processed
```

```python
# Check for missing values
print(data.isna().sum().sum())

# Replace '?' with NaN
data.replace('?', np.nan, inplace=True)
print(data.isna().sum().sum())

# Drop rows with missing values
data.dropna(inplace=True)
print(data.isna().sum().sum())

# Summarize Dataset
data.describe(include='all')
```

```python
# Data set attributes
age = data.iloc[:, 0]
capital_gain = data['capital-gain']
capital_loss = data['capital-loss']
hours_per_week = data['hours-per-week']

# Drop unnecessary columns
data.drop(['fnlwgt', 'education-num', 'relationship', 'occupation'], axis=1, inplace=True)

# Summarize Dataset
data.describe(include='all')
```

```python
# Preprocessing


# Encoding Income columns
data['income'] = data['income'].map({' <=50K': 'less', ' >50K': 'more'})
# Discretization of numerical continuous variables into categorical attributes
data['age'] = pd.qcut(data['age'], q=10, labels=False, duplicates='drop')
data['capital-gain'] = pd.cut(data['capital-gain'], bins=2, labels=False)
data['capital-loss'] = pd.cut(data['capital-loss'], bins=3, labels=False)
data['hours-per-week'] = pd.qcut(data['hours-per-week'], q=2, labels=False)

# Convert all columns to string type
data = data.astype(str)
# one-hot encoding
data_processed = pd.get_dummies(data)
# Binarize the dataset (convert non-zero values to 1)
data_processed[data_processed != 0] = 1
data_processed
```

```python
#a

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
from IPython.display import display, HTML

# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Rank top 10 association rules for different combinations of support and confidence
support_confidence_combinations = [(0.1, 0.7), (0.2, 0.8), (0.3, 0.9)]

for support, confidence in support_confidence_combinations:
    print(f"\nTop 10 association rules with support={support} and confidence={confidence}:\n")
    filtered_rules = rules[(rules['support'] >= support) & (rules['confidence'] >= confidence)]
    top_10_rules = filtered_rules.sort_values(by='lift', ascending=False).head(10)
    display(HTML(top_10_rules.to_html()))
```

```python
#b

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Define a function to calculate interest
def calculate_interest(rule):
    return rule.confidence - rule.support / len(data_processed)

# Calculate interest for each rule
rules['interest'] = rules.apply(calculate_interest, axis=1)

# Rank top 10 association rules for different combinations of support and interest
support_interest_combinations = [(0.1, 0.2), (0.2, 0.3), (0.3, 0.4)]

for support, interest in support_interest_combinations:
    print(f"\nTop 10 association rules with support={support} and interest={interest}:\n")
    filtered_rules = rules[(rules['support'] >= support) & (rules['interest'] >= interest)]
    top_10_rules = filtered_rules.sort_values(by='lift', ascending=False).head(10)
    display(HTML(top_10_rules.to_html()))
```

```python
import numpy as np
# Perform one-hot encoding and find frequent itemsets
frequent_itemsets = apriori(data_processed, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# The J-Measure is a measure of the worth of an implication rule A -> B. It's defined as:

# J(A -> B) = P(A, B) * log2(P(B|A) / P(B)) + P(A, B) * log2(P(B|A) / P(~B))

# Calculate probabilities
rules['P(A, B)'] = rules['support']
rules['P(B|A)'] = rules['confidence']
rules['P(A, ~B)'] = rules['antecedent support'] - rules['support']
rules['P(~B|A)'] = 1 - rules['confidence']
rules['P(B)'] = rules['consequent support']
rules['P(~B)'] = 1 - rules['consequent support']

# Calculate J-Measure
rules['J-Measure'] = rules['P(A, B)'] * np.log2(rules['P(B|A)'] / rules['P(B)']) + rules['P(A, ~B)'] * np.log2(rules['P(~B|A)'] / rules['P(~B)'])

# Show top 10 rules with highest J-Measure
rules = rules.sort_values(by='J-Measure', ascending=False)
print("Top 10 rules with highest J-Measure:")
rules.head(10)
```
